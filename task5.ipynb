```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:/Users/khasi/OneDrive/Desktop/elevate labs/student_habits_performance.csv")
df
```

```python
# Consider only the first 98 entries
df = df.head(98)
df
```

```python
# Basic Info
print("Data Info:")
print(df.info())
```

```python
# Summary Statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Value Counts for categorical columns
for col in df.select_dtypes(include='object').columns:
    print(f"\nValue counts for {col}:\n", df[col].value_counts())
```

```python
# Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Correlation Heatmap (only numeric columns)
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Histograms
df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Histograms of Features')
plt.show()
```

```python
# Boxplots
plt.figure(figsize=(12,8))
sns.boxplot(data=df)
plt.title('Boxplots of Numeric Features')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatterplot Example
plt.figure(figsize=(6,4))
sns.scatterplot(x='study_hours_per_day', y='exam_score', data=df)
plt.title('Study Hours vs Performance Score')
plt.show()
```
